```python
import xgboost as xgb
import pickle
import pandas as pd
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score
from keras.models import load_model
```

```python
# word2vec: skip-gram
X_train = pd.read_csv('../Word Embedding/emb_sg_train.csv')
X_test = pd.read_csv('.../Word Embedding/emb_sg_test.csv')

# labels
y_train = pd.read_csv('../Final Data - Test train/y_train.csv')['class']
y_test = pd.read_csv('../Final Data - Test train/y_test.csv')['class']
```

```python
%%time
xgb_model = pickle.load(open('../Models/xgb.pkl', "rb"))
xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
```

```python
%%time
model = load_model('../Models/CNN.h5')
model.fit(X_train, y_train)
y_pred_class = list(map(lambda x: 1 if x > 0.5 else 0, y_pred))
accuracy = accuracy_score(y_test, y_pred_class)
f1 = f1_score(y_test, y_pred_class, average='weighted')
precision = precision_score(y_test, y_pred_class)
recall = recall_score(y_test, y_pred_class)
```

```python
metrics = pd.DataFrame().append({'Model': 'CNN', 'Accuracy': accuracy, 'F1-score': f1, 'Precision': precision, 'Recall': recall}, ignore_index = True)
metrics
```
